```python
from sklearn import tree, metrics
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
matplotlib.rc("xtick", labelsize=20)
matplotlib.rc("xtick", labelsize=20)
```

```python
df= pd.read_csv("winequality-red.csv")
```

```python
df
```

```python
#changing the scale of quality in the dataset from 3-8 to 0-5
quality_mapping={3:0, 4:1, 5:2, 6:3, 7:4, 8:5 }
df.loc[:,"quality"]=df["quality"].map(quality_mapping)
```

```python
#shuffle the dataset and create training set and validation set
df = df.sample(frac=1).reset_index(drop=True)
df_train = df.head(1000)
df_val = df.tail(599)
```

```python

train_acclist=[]
val_acclist=[]
cols=['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 
      'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']
for depth in range(1,25):
    clf = tree.DecisionTreeClassifier(max_depth = depth)
    clf.fit(df_train[cols],df_train['quality'])
    train_pred = clf.predict(df_train[cols])
    val_pred = clf.predict(df_val[cols])
    train_acc = metrics.accuracy_score(df_train['quality'], train_pred)
    val_acc = metrics.accuracy_score(df_val['quality'], val_pred)
    train_acclist.append(train_acc)
    val_acclist.append(val_acc)
    
```

```python
# plotting train vs val accuracy so we can understan how overfiting works
plt.figure(figsize = (10,5))
sns.set_style("whitegrid")
plt.plot(train_acclist, label = "train accuracy")
plt.plot(val_acclist, label = "validation accuracy")
plt.legend(loc = "upper left", prop = {'size':15})
plt.xticks(range(0,26,5))
plt.xlabel("max_depth", size = 20)
plt.ylabel("accuracy", size = 20)
plt.show()
```
